```python
import pandas as pd

# Di Colab, file yang diupload langsung masuk ke folder root '/content/'
train = pd.read_csv('/home/haikal/Downloads/house-pricing/house-prices-advanced-regression-techniques/train.csv')
train
```

```python
# Memeriksa jumlah nilai yang hilang di setiap kolom
missing_values = train.isnull().sum()
missing_values[missing_values > 0]
```

```python
#melihat dan menangani missing value
over = missing_values[missing_values >1000].index
# over
less = missing_values[missing_values <1000].index
less
#dari idata di atas di ketahui bahwa missing valuenya kurang dari 75% sehingga penanganan missing valuenya kita menggantinya bukan menghapusnya
```

```python
# Contoh mengisi nilai yang hilang dengan median untuk kolom numerik
numeric_features = train[less].select_dtypes(include=['number']).columns
train[numeric_features] = train[numeric_features].fillna(train[numeric_features].median())
```

```python
# Contoh mengisi nilai yang hilang dengan mode untuk kolom kategori
kategorical_features = train[less].select_dtypes(include=['str']).columns

for column in kategorical_features:
    train[column] = train[column].fillna(train[column].mode()[0])
```

```python
# Menghapus kolom dengan terlalu banyak nilai yang hilang
df = train.drop(columns=over)
```

```python
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
 
for feature in numeric_features:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[feature])
    plt.title(f'Box Plot of {feature}')
    plt.show()
```

setelah kita melihat dari box plot di atas banyak terdapat outler di setiap fitur

```python
# Contoh sederhana untuk mengidentifikasi outliers menggunakan IQR
Q1 = df[numeric_features].quantile(0.25)
Q3 = df[numeric_features].quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
```

```python
# 1. Mengganti nilai outlier dengan batas bawah dan atas
df[numeric_features] = df[numeric_features].clip(lower=lower_limit, upper=upper_limit, axis=1)

# 2. Karena tidak ada baris yang dihapus, df tetap utuh (numerik & kategorikal)
print("Jumlah baris tetap:", df.shape[0])
```

```python
# Menghitung jumlah variabel
num_vars = df_one_hot.shape[1]
 
# Menentukan jumlah baris dan kolom untuk grid subplot
n_cols = 4  # Jumlah kolom yang diinginkan
n_rows = -(-num_vars // n_cols)  # Ceiling division untuk menentukan jumlah baris
 
# Membuat subplot
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, n_rows * 4))
 
# Flatten axes array untuk memudahkan iterasi jika diperlukan
axes = axes.flatten()
 
# Plot setiap variabel
for i, column in enumerate(df_one_hot.columns[:20]):  # Membatasi hanya pada 20 variabel pertama untuk visualisasi
    df_one_hot[column].hist(ax=axes[i], bins=20, edgecolor='black')
    axes[i].set_title(column)
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Frequency')
 
# Menghapus subplot yang tidak terpakai (jika ada)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])
 
# Menyesuaikan layout agar lebih rapi
plt.tight_layout()
plt.show()
```
